1. Use matplotlib to plot the following equation:
y = x^2 - x + 2
You'll need to write the code that generates the x and y points.

    Add an anotation for the point 0, 0, the origin.

```python
import matplotlib.pyplot as plt
import numpy as np

x = list(range(-50,50))
y = [(n ** 2) - n + 2 for n in x]
plt.plot(x, y, color = 'bisque')

plt.annotate('(0,0)', (0,0), fontsize = 10, color = 'blue')
# plt.text(0, 0, '(0, 0)', fontsize=10, color='blue')

plt.title('$y = x^2 - x + 2$')
plt.show()
```

3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize = (9,9))


plt.subplot(221)
x = list(range(100))
y = [math.sqrt(n) for n in x]
plt.plot(x, y, color = '#f0ead6')
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = sqrt(x)')

plt.subplot(222)
x = list(range(100))
y = [n ** 3 for n in x]
plt.plot(x, y, color = '#a0785a')
plt.xlabel('x')
plt.ylabel('y')
plt.title('$y = x^3$')

plt.subplot(223)
x = list(range(0,51))
y = [2**n for n in x]
plt.plot(x, y, color = '#644117')
plt.xlabel('x')
plt.ylabel('y')
plt.title('$y = 2^x$')

plt.subplot(224)
x = list(range(0,51))
y = [1/(n+1) for n in x]
plt.plot(x, y, color = '#c19a6b')
plt.xlabel('x')
plt.ylabel('y')
plt.title('$y = 1/(x+1)$')

plt.tight_layout()
```

4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
plt.figure(figsize = (9,9))
plt.suptitle('Math Equations')

# plot1
x = list(range(100))
y = [math.sqrt(n) for n in x]
plt.plot(x, y, color = '#8d5524', label = 'y = \sqrt(x)')
plt.xlabel('x')
plt.ylabel('y')

# plot2
x = list(range(100))
y = [n ** 3 for n in x]
plt.plot(x, y, color = '#c68642', label = '$y = x^3$')
plt.xlabel('x')
plt.ylabel('y')

# plot3
x = list(range(100))
y = [2**n for n in x]
plt.plot(x, y, color = '#e0ac69', label = '$y = 2^x$')

# plot 4
x = list(range(0,100))
y = [1/(n+1) for n in x]
plt.plot(x, y, color = '#f1c27d', label = '$y = 1/(x+1)$')

plt.legend(loc='upper right')
plt.tight_layout()
```
