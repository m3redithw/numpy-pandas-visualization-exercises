1. Use matplotlib to plot the following equation:
$y = x^2 - x + 2$

    You'll need to write the code that generates the x and y points.

    Add an anotation for the point 0, 0, the origin.

```python
import matplotlib.pyplot as plt
import numpy as np

# generate data
x = list(range(-50,50))
y = [(n ** 2) - n + 2 for n in x]

# plot x and y
plt.plot(x, y, color = 'bisque', linewidth = 5)

# add lables
plt.xlabel('$x$')
plt.ylabel('$y$')

# add annotation
plt.annotate('(0,0)', (0,0), fontsize = 10, color = 'blue')
## plt.text(0, 0, '(0, 0)', fontsize=10, color='blue')

# add title
plt.title('$y = x^2 - x + 2$')
plt.show()

# %config InlineBackend.figure_format = 'svg'

# save the figure
plt.savefig('myfirstfigure.png')
```

2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
- $y = âˆšx$
- $y = x^3$
- $y = 2^x$
- $y = 1/(x+1)$

You can use functions from the `math` module to help implement some of the equations above.

```python
# y = \sqrt(x)
import math
x = list(range(100))
y = [math.sqrt(n) for n in x]
plt.plot(x, y, color = '#f0ead6', linewidth = 5)
plt.grid(True, ls = '--')
```

```python
# y = x^3
x = list(range(100))
y = [n ** 3 for n in x]
plt.plot(x, y, color = '#f1c27d', linewidth = 5)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, ls = '--')
```

```python
# y = 2^x
x = list(range(0,51))
y = [2**n for n in x]
plt.plot(x, y, c = '#e0ac69', linewidth = 5)
plt.grid(True, ls = '--')
```

```python
# y = 1/(x+1)
x = list(range(0,51))
y = [1/(n+1) for n in x]
plt.plot(x, y, c = '#a0785a', linewidth = 5)
plt.grid(True, ls = '--')
```

3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize = (9,9))
plt.suptitle('Math Equations')

# subplot 1
plt.subplot(221)
x = list(range(100))
y = [math.sqrt(n) for n in x]
plt.plot(x, y, color = '#f0ead6', linewidth = 5)
plt.xlabel('x')
plt.ylabel('$\sqrt{x}$')
plt.title('$y = \sqrt{x}$')

# subplot 2
plt.subplot(222)
x = list(range(100))
y = [n ** 3 for n in x]
plt.plot(x, y, color = '#f1c27d', linewidth = 5)
plt.xlabel('x')
plt.ylabel('$x^3$')
plt.title('$y = x^3$')

# subplot 3
plt.subplot(223)
x = list(range(0,51))
y = [2**n for n in x]
plt.plot(x, y, color = '#c68642', linewidth = 5)
plt.xlabel('x')
plt.ylabel('$2^x$')
plt.title('$y = 2^x$')

# subplot 4
plt.subplot(224)
x = list(range(0,51))
y = [1/(n+1) for n in x]
plt.plot(x, y, color = '#644117', linewidth = 5)
plt.xlabel('x')
plt.ylabel('$1/(x+1)$')
plt.title('$y = 1/(x+1)$')

plt.tight_layout()
```

4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
plt.figure(figsize = (9,9))
plt.suptitle('Math Equations')

plt.xlabel('x')
plt.ylabel('y')

# plot1
x = list(range(100))
y = [math.sqrt(n) for n in x]
plt.plot(x, y, color = '#8d5524', label = '$y = \sqrt{x}$', linewidth = 3)


# plot2
x = list(range(100))
y = [n ** 3 for n in x]
plt.plot(x, y, color = '#c68642', label = '$y = x^3$', linewidth = 3)


# plot3
x = list(range(100))
y = [2**n for n in x]
plt.plot(x, y, color = '#e0ac69', label = '$y = 2^x$', linewidth = 3)

# plot 4
x = list(range(0,100))
y = [1/(n+1) for n in x]
plt.plot(x, y, color = '#f1c27d', label = '$y = 1/(x+1)$', linewidth = 3)

plt.legend(loc='upper right')
plt.ylim(0,10)
plt.tight_layout()
```
