- Label your x axis "Elements"
- Label your y axis "Operations"
- Label your curves or make a legend for the curves
- Use LaTex notation where possible

```python
import matplotlib.pyplot as plt
import math
plt.figure(figsize = (9,9))
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.suptitle('Big O Notation')

# y = 0n +1
n = range(-50,51)
y = [0*n + 1 for n in n]
plt.plot(n, y, c = '#ff6961', label = '$O(1)$')

# y = log(n)
n = range(1, 51)
y = [math.log(n)for n in n]
plt.plot(n, y, c = '#ffb480', label = '$O(log n)$')

# y = n
n = range(-50,51)
y = [n for n in n]
plt.plot(n, y, c = '#f8f38d', label = '$O(n)$')

# y = n * log(n)
n = range(1,51)
y = [n * math.log(n) for n in n]
plt.plot(n, y, c = '#42d6a4', label = '$O(n log n)$')

# y = n^2
n = range(-50, 51)
y = [n ** 2 for n in n]
plt.plot(n, y, c = '#08cad1', label = '$O(n^2)$')

# y = 2^n
n = range(-50,51)
y = [2 ** n for n in n]
plt.plot(n, y, c = '#59adf6', label = '$O(2^n)$')

# y = n!
n = range(0,51)
y = [math.factorial(n) for n in n]
plt.plot(n, y, c = '#9d94ff', label = '$O(n!)$')

# y = n^n
n = range(-50,51)
y = [n ** n for n in n]
plt.plot(n, y, c = '#c780e8', label = '$O(n^n)$')

plt.xlim(1,11)
plt.ylim(0,50)
plt.legend(loc = 'upper right')
plt.title('Meredith Wang', size = 10)
plt.tight_layout()
```
